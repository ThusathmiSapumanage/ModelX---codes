```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Function to output outliers in the column that we input using IQR
def find_outliers_IQR(df_cleaned):
  q1=df_cleaned.quantile(0.25)
  q3=df_cleaned.quantile(0.75)
  IQR=q3-q1
  outliers = df_cleaned[((df_cleaned<(q1-1.5*IQR))|(df_cleaned>(q3+1.5*IQR)))]
  return outliers
```
